↓ importing libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

↓ code for web scraping the young adult page of Books to Scrape

```python
base_url = "https://books.toscrape.com/catalogue/category/books/young-adult_21/index.html" # base url for the young adult book category 

books = []  # initializes an empty list to store book data
page = 1 

while True:  # loops to go through the pages in the mystery category
    if page > 1:
        url = base_url.replace("index.html", f"page-{page}.html")  # updates url for the following pages
    else:
        url = base_url  # uses the base url for the first page

    response = requests.get(url) # sends an HTTP GET request to the url
    if response.status_code != 200:  # breaks if the page doesn't exist
        break

    soup = BeautifulSoup(response.text, "html.parser") # parses the HTML content with BeautifulSoup
    
    # loops through each book listing in the page
    for book in soup.select("article.product_pod"):  # iterates over each book entry in the product listing
        title = book.h3.a["title"]  # extracts the book's title
        price = book.select_one("p.price_color").text  # extracts the book's price
        rating = book.select_one("p.star-rating")["class"][1]  # extracts the book's rating
        availability = book.select_one("p.instock.availability").text.strip()  # extracts the availability status
        
        # appends a dictionary with book data to the list
        books.append({
            "Title": title,
            "Price": price,
            "Rating": rating,
            "Availability": availability,
        })
        
    page += 1  

df = pd.DataFrame(books)  # converts the books list to a DataFrame for easy manipulation
df['Price'] = df['Price'].str.replace('Â', '', regex=False)  # used ChatGPT, for data cleaning purposes
df.to_csv("young_adult_books.csv", encoding='utf-8', index=False)  # saves the DataFrame to a CSV file without the index
print("Data has been saved.")  # prints confirmation that the data has been saved
```

#### ———  code for analysis

#### Question 1: Among (combining) the five categories, which price range is the most common?

```python
# loads the data from the CSV files
df_young = pd.read_csv("young_adult_books.csv")
df_mystery = pd.read_csv("mystery_books.csv")
df_historical = pd.read_csv("historical_fiction_books.csv")
df_sequential = pd.read_csv("sequential_art_books.csv")
df_romance = pd.read_csv("romance_books.csv")

df_all = pd.concat([df_young, df_mystery, df_historical, df_sequential, df_romance]) # combines all the dataframes into one

df_all['Price'] = pd.to_numeric(df_all['Price'].str.replace('£', ''), errors='coerce') # used ChatGPT for this, converts the price to numeric values

price_intervals = [0, 10, 20, 30, 40, 50, 100] # defines intervals for the price ranges
price_labels = ['£0-10', '£11-20', '£21-30', '£31-40', '£41-50', '£50+'] # defines labels for each range
df_all['Price Range'] = pd.cut(df_all['Price'], bins=price_intervals, labels=price_labels, right=False) # categorizes prices into defined ranges, right=False ensures that the upper boundary is excluded for each range

plt.figure(figsize=(10, 6))  # sets the size of the plot
sns.set(style="whitegrid")  # sets the styling for the plot
price_plot = sns.countplot(data=df_all, x='Price Range', hue='Price Range', order=price_labels, palette='PuRd', dodge=False)  # creates a count plot for the price intervals

price_plot.set_title("The Most Common Price Range of Books Across All Five Categories Is £11-20")  # sets the title for the plot
price_plot.set_ylabel("Number of Books")  # sets the y-axis label                        
plt.show()  # displays the plot
```

#### Question 2: What category has the most books that are a part of a series?

```python
# loads the data from the CSV files
df_young = pd.read_csv("young_adult_books.csv")
df_mystery = pd.read_csv("mystery_books.csv")
df_historical = pd.read_csv("historical_fiction_books.csv")
df_sequential = pd.read_csv("sequential_art_books.csv")
df_romance = pd.read_csv("romance_books.csv")

# add a category column for each datafeame
df_young['Category'] = 'Young Adult'
df_mystery['Category'] = 'Mystery'
df_historical['Category'] = 'Historical Fiction'
df_sequential['Category'] = 'Sequential Art'
df_romance['Category'] = 'Romance'

df_all = pd.concat([df_young, df_mystery, df_historical, df_sequential, df_romance]) # combines all the dataframes into one

df_all['In_Series'] = df_all['Title'].str.contains(r'\(.*\)') # create a new column to identify books that are part of a series (contains parentheses in the title)

series_books = df_all[df_all['In_Series']] # filter the dataframe to include only books that are part of a series

category_counts = series_books['Category'].value_counts().reset_index() # groups by category and count the number of books in each category that are part of a series
category_counts.columns = ['Category', 'Count'] # renames the columns

plt.figure(figsize=(10, 6)) # sets the size of the plot
sns.barplot(data=category_counts, x='Category', y='Count', palette='PuRd', hue='Category') # creates a bar plot showing the count of series books for each category

plt.title('Sequential Art Has the Greatest Number of Books That Are a Part of a Series') # sets the title for the plot
plt.ylabel('Number of Books') # sets the y-axis label  
plt.xlabel('') # sets the x-axis label to an empty string for a cleaner look
plt.show() # displays the plot
```

#### Question 3: Out of the five categories, what catagory has the highest average rating?

```python
# loads the data from the CSV files
df_young = pd.read_csv("young_adult_books.csv")
df_mystery = pd.read_csv("mystery_books.csv")
df_historical = pd.read_csv("historical_fiction_books.csv")
df_sequential = pd.read_csv("sequential_art_books.csv")
df_romance = pd.read_csv("romance_books.csv")

rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}  # creates a dictionary to map rating words to numbers

# applies the rating map to each category
df_young['Rating'] = df_young['Rating'].map(rating_map)
df_mystery['Rating'] = df_mystery['Rating'].map(rating_map)
df_historical['Rating'] = df_historical['Rating'].map(rating_map)
df_sequential['Rating'] = df_sequential['Rating'].map(rating_map)
df_romance['Rating'] = df_romance['Rating'].map(rating_map)

# calculates the average rating for each category
avg_ratings = {
    "Young Adult": df_young['Rating'].mean(),
    "Mystery": df_mystery['Rating'].mean(),
    "Historical Fiction": df_historical['Rating'].mean(),
    "Sequential Art": df_sequential['Rating'].mean(),
    "Romance": df_romance['Rating'].mean()
}

# finds the category with the highest average rating
highest_avg_rating_category = max(avg_ratings, key=avg_ratings.get) # finds the category with the highest average rating from the avg_ratings dictionary
highest_avg_rating = avg_ratings[highest_avg_rating_category] # gets the highest average rating value associated with the highest-rated category
avg_ratings_df = pd.DataFrame(list(avg_ratings.items()), columns=['Category', 'Average Rating']) # converts the avg_ratings dictionary into a dataframe

plt.figure(figsize=(10, 6)) # sets the size of the plot
sns.barplot(data=avg_ratings_df, x='Category', y='Average Rating', palette='PuRd', hue='Category') # creates a bar plot showing the count of series books for each category

plt.title('Out Of The Five Catagories, Young Adult Has The Highest Rating') # sets the title for the plot
plt.ylabel('Average Rating') # sets the y-axis label  
plt.xlabel('') # sets the x-axis label to an empty string for a cleaner look
plt.show() # displays the plot
```

#### Question 4: Does having more books in a category correlate with a higher average rating?

```python
# loads the data from the CSV files
df_young = pd.read_csv("young_adult_books.csv")
df_mystery = pd.read_csv("mystery_books.csv")
df_historical = pd.read_csv("historical_fiction_books.csv")
df_sequential = pd.read_csv("sequential_art_books.csv")
df_romance = pd.read_csv("romance_books.csv")

rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}  # creates a dictionary to map rating words to numbers

# applies the rating map to each category
df_young['Rating'] = df_young['Rating'].map(rating_map)
df_mystery['Rating'] = df_mystery['Rating'].map(rating_map)
df_historical['Rating'] = df_historical['Rating'].map(rating_map)
df_sequential['Rating'] = df_sequential['Rating'].map(rating_map)
df_romance['Rating'] = df_romance['Rating'].map(rating_map)

# calculates the average rating for each category
avg_ratings = {
    "Young Adult": df_young['Rating'].mean(),
    "Mystery": df_mystery['Rating'].mean(),
    "Historical Fiction": df_historical['Rating'].mean(),
    "Sequential Art": df_sequential['Rating'].mean(),
    "Romance": df_romance['Rating'].mean()
}

# counts the number of books in each category
book_counts = {
    "Young Adult": len(df_young),
    "Mystery": len(df_mystery),
    "Historical Fiction": len(df_historical),
    "Sequential Art": len(df_sequential),
    "Romance": len(df_romance)
}

# creates a dataframe with both book counts and average ratings
analysis_df = pd.DataFrame({
    'Category': list(avg_ratings.keys()),
    'Book Count': list(book_counts.values()),
    'Average Rating': list(avg_ratings.values())
})


plt.figure(figsize=(10, 6))  # set the size of the plot
sns.barplot(data=analysis_df, x='Book Count', y='Average Rating', palette='PuRd', hue='Category')  # create a bar plot with book count on x and average rating on y
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # makes a legend

plt.title('The Number of Books in a Category Does Not Strongly Correlate with Higher Average Ratings')  # sets the title of the plot
plt.xlabel('Number of Books')  # sets the x-axis label
plt.ylabel('Average Rating')  # sets the y-axis label
plt.show()  # displays the plot
```
